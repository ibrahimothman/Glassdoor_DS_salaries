```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('all_data_cleaned_20_01_2022.csv')
```

```python
df = df.drop(['Unnamed: 0'], axis=1)
```

```python
# let's look at some correlations
cmap = sns.diverging_palette(230, 20, as_cmap=True)
corr = df[['c_age', 's_avg', 'Rating', 'desc_len']].corr()
# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

Let's check out the categorical variables

```python
df_cat = df[['Location', 'Size', 'Type of ownership', 'Industry', 'Sector', 'Revenue', 'c_name', 'state', 'py_req', 'r_req', 'excel_req', 'sql_req', 'tableau_req','seniority']]
```

```python
for col in df_cat.columns:
    cat_counts = df_cat[col].value_counts()
    print(f'The graph for {col} - total = {len(cat_counts)}')
    sns.barplot(x = cat_counts.index, y = cat_counts, data = df_cat)
    plt.xticks(rotation=90)
    plt.show()
    
```

```python
for col in df_cat[['Location', 'c_name']].columns:
    cat_counts = df_cat[col].value_counts()[:20]
    print(f'The graph for {col} - total = {len(cat_counts)}')
    sns.barplot(x = cat_counts.index, y = cat_counts, data = df_cat)
    plt.xticks(rotation=90)
    plt.show()
    
```

```python
# let's create some pivot tables for salaries
pd.pivot_table(df, index= 'job_simp', values=['s_avg'])
```

```python
pd.pivot_table(df, index= ['job_simp', 'seniority'], values=['s_avg'])
```

```python
pd.pivot_table(df, index= ['state', 'job_simp'], values=['s_avg'], aggfunc='count').sort_values('state', ascending=False)
```

```python
pd.pivot_table(df[df.job_simp == 'data scientist'], index= ['state'], values=['s_avg']).sort_values('s_avg', ascending=False)
```

```python
# check out salary VS other variables
df_pivots = df[['desc_len', 'Rating', 'Size', 'Type of ownership', 'Sector', 'c_age', 'Industry',
       'Revenue', 'py_req', 'r_req', 'excel_req', 'sql_req', 'tableau_req',
       'seniority']]
```

```python
for col in df_pivots.columns:
    print(pd.pivot_table(df, index = col, values=['s_avg']).sort_values('s_avg', ascending=False))
```
